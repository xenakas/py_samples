# Австралийские аэропорты

```python
import numpy as np
import pandas as pd
```

```python
airports = pd.read_csv('airports.csv',names = ['Airport ID','Name','City','Country','IATA','ICAO','Latitude','Longitude','Altitude','Timezone','DST','Tz_database_time_zone','Type','Source'])
routes = pd.read_csv('routes.csv',names = ['Airline','Airport ID','Source airport','Source airport ID','Destination airport','Destination airport ID','Codeshare','Stops','Equipment'])

airports = airports.replace('\\N', np.nan)
routes = routes.replace('\\N', np.nan)

airports = airports.dropna(subset=['IATA']) # из датасета удаляется 36 австралийских аэропортов, можно было бы заменить пропущенные коды IATA на ICAO без первой буквы, но эти значения совпадают для очень малого числа аэропортов 
```

Выберем австралийские аэропорты

```python
aus_airports = airports[airports['Country'].isin(['Australia'])]
#aus_airports[airports['IATA'].isnull()]
```

Выберем маршруты из городов Австралии в города других стран

```python
aus_list = aus_airports['IATA'].unique()
aus_routes = routes[(routes['Source airport'].isin(aus_list) ) &  (~routes['Destination airport'].isin(aus_list) ) ]
```

Посчитаем число маршрутов для каждого аэропорта

```python
aus_routes_stacked = aus_routes.groupby(['Source airport', 'Destination airport' ]).count()['Airline'].unstack(level=0)
aus_routes_stacked['IATA'] = aus_routes_stacked.index
```

```python
result = aus_routes_stacked.merge(airports[['IATA','Country', 'City', 'Name']],left_on='IATA', right_on='IATA' , how='inner' )
del result['IATA']
```

```python
result = result.groupby(['Country',  'Name']).sum().rename(index=str,columns={'ADL': 'Adelaide International Airport', 'BNE': 'Brisbane International Airport', 'CNS': 'Cairns International Airport',  'DRW': 'Darwin International Airport', 'MEL': 'Melbourne International Airport', 'OOL': 'Gold Coast Airport', 'PER': 'Perth International Airport', 'PHE': 'Port Hedland International Airport', 'SYD': 'Sydney Kingsford Smith International Airport'})
```
