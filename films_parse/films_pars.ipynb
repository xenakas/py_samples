```python
soup = BeautifulSoup(r.text, 'html.parser')
n_page = int(soup.find_all('a',{'class':'paginator__page-number'})[-1].text)
```

```python
for n in tqdm.tqdm_notebook(range(1,n_page+1)):
    r = s.get(f'https://www.kinopoisk.ru/lists/navigator/country-2/?page={n}&quick_filters=films&tab=all')

    soup = BeautifulSoup(r.text, 'html.parser')
    
    active = soup.find_all('div',{'class':'desktop-seo-selection-film-item selection-list__film'})
    for i in range(len(active)):
        data = {}
        try:
            data['Rating'] = [float(active[i].find('span',{'class':'rating__value rating__value_positive'}).text)]
        except:
            data['Rating'] = None
            
        try:
            data['Rating_N_pers'] = [int(active[i].find('span',{'class':'rating__count'}).text.replace(' ',''))]
        except:
            data['Rating_N_pers'] = None
            
        try:
            data['Name'] = [active[i].find('p',{'class':'selection-film-item-meta__name'}).text]
        except:
            data['Name'] = None
            
        try:
            data['Genre'] =[ active[i].find_all('span',{'class':'selection-film-item-meta__meta-additional-item'})[-1].text]
        except:
            data['Genre'] = None
            
        try:
            data['Country'] = [active[i].find('span',{'class':'selection-film-item-meta__meta-additional-item'}).text]
        except:
            data['Country'] = None
        try:
            data['Year'] = [active[i].find('p',{'class':'selection-film-item-meta__original-name'}).text]
        except:
            data['Year'] = None
        try:
            data['href'] = ['https://www.kinopoisk.ru'+active[i].find('a',{'class':'selection-film-item-meta__link'}).attrs['href']]
        except:
            data['href'] = None
        

        
        full_data = pd.concat([full_data,pd.DataFrame(data)])
        print(data)
    time.sleep(5)
```
