#### Loop over DF

Iterating over a Pandas DataFrame is typically done with the iterrows() method

```python
for lab, row in cars[3:5].iterrows():
    print(lab)
    print(row)
    print(str(lab) + ": " + str(row["cars_per_cap"]))
```

```python
# Code for loop that adds COUNTRY column
for lab, row in cars.iterrows():
    cars.loc[lab, 'random'] = npr.randint(0, 20)
    
print(cars.head(4))
```

Using iterrows() to iterate over every observation of a Pandas DataFrame is easy to understand, but not very efficient. On every iteration, you're creating a new Pandas Series.

If you want to add a column to a DataFrame by calling a function on another column, the iterrows() method in combination with a for loop is not the preferred way to go. Instead, you'll want to use apply().

```python
cars['color'] = npr.choice(colors, np.shape(cars)[0])

print(cars.head(4))
```

```python
cars['COLOR'] = cars['color'].apply(str.upper)

print(cars.head(4))
```

 ## Logic, Control Flow and Filtering

```python
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house > 18.5,  my_house < 10)
)
# Both my_house and your_house smaller than 11
print( np.logical_and(my_house <11, your_house <11))
```

If you're dealing with a 2D Numpy array, it's more complicated. A 2D array is built up of multiple 1D arrays. To explicitly iterate over all separate elements of a multi-dimensional array, you'll need this syntax:

#### Random Walk

```python
# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = max(0, step - 1) #make sure that a variable x never goes below 0:
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)

plt.plot(random_walk)
```

```python

# Initialize all_walks
all_walks = []

# Simulate random walk 10 times
for i in range(100) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        
        if np.random.rand() <= 0.001  : # 0.1% chance of falling down. That calls for another random number generation. Basically, you can generate a random float between 0 and 1. If this value is less than or equal to 0.001, you should reset step to 0.
            step = 0

        
        random_walk.append(step)
        
        
    # Append random_walk to all_walks
    all_walks.append(random_walk)

    
plt.plot(np.transpose(np.array( all_walks )))
plt.show()

plt.hist(np.transpose(np.array( all_walks ))[-1], bins = 35)
plt.show()
```
