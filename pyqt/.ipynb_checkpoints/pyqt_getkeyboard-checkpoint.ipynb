### Обработка нажатий клавиатуры и мыши

[Все имена кнопок и их коды](https://doc.qt.io/qt-5/qt.html)

```python
import sys
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QWidget, QApplication, QLabel, QInputDialog, QPushButton, QColorDialog
from PyQt5.QtGui import QPainter, QColor
 
 
class Example(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
 
    def initUI(self):
        self.setGeometry(300, 300, 300, 300)
        self.setWindowTitle('pyqtgraph')
 
        self.coords = QLabel(self)
        self.coords.setText("Drag coordinates:0, 0           ")
        self.coords.move(30, 30)
        
        self.setMouseTracking(True)
    
        self.arrows = QLabel(self)
        self.arrows.setText("Arrow direction: None      ")
        self.arrows.move(30, 60)
             
        
        self.button_1 = QPushButton(self)
        self.button_1.move(30, 150)
        self.button_1.setText("Get name")
        self.button_1.clicked.connect(self.run)

        self.button_2 = QPushButton(self)
        self.button_2.move(30, 180)
        self.button_2.setText("Get age")
        self.button_2.clicked.connect(self.run1)

        self.button_3 = QPushButton(self)
        self.button_3.move(30, 210)
        self.button_3.setText("Get country")
        self.button_3.clicked.connect(self.run2)

        self.button_4 = QPushButton(self)
        self.button_4.move(30, 240)
        self.button_4.setText("Get color")
        self.button_4.clicked.connect(self.run3)
 
        
        self.show()
 
    def mouseMoveEvent(self, event):
        self.coords.setText("Drag coordinates:{}, {}".format(event.x(), event.y()))
   
        
    def keyPressEvent(self, event):
        if event.key() == Qt.Key_Up:
            self.arrows.setText("Arrow direction: Up")
        elif event.key() == Qt.Key_Down:
            self.arrows.setText("Arrow direction: Down")
        elif event.key() == Qt.Key_Right:
            self.arrows.setText("Arrow direction: Right")
        elif event.key() == Qt.Key_Left:
            self.arrows.setText("Arrow direction: Left")
        else: 
            self.arrows.setText("Arrow direction: None")
            
        if int(event.modifiers()) == (Qt.AltModifier + Qt.ShiftModifier):
            if event.key() == Qt.Key_B:
                self.arrows.setText("Hi, there")

    def paintEvent(self, event):
        qp = QPainter()
        qp.begin(self)
        self.drawFlag(qp)
        qp.end()        
            
    def drawFlag(self,qp):
        qp.setBrush(QColor(25, 46, 200))
        qp.drawRect(30, 90, 30, 30)
        qp.setBrush(QColor(44, 255, 100))
        qp.drawRect(75, 90, 30, 30)
        qp.setBrush(QColor(150, 11, 125))
        qp.drawRect(120, 90, 30, 30)
    
    def run(self):
        i, okBtnPressed = QInputDialog.getText(
            self, "Enter name", "What's up?")
        if okBtnPressed:
            self.button_1.setText("Thx, bro")
            print(i)
            print(okBtnPressed)
    
    def run1(self):
        i, okBtnPressed = QInputDialog.getInt(
            self, "Enter age", "What's up?", 21,18,99,1) # ..., current, min, max, gap
        if okBtnPressed:
            self.button_2.setText("Thx, bro")
            print(i)
            print(okBtnPressed)
 
    def run2(self):
        i, okBtnPressed = QInputDialog.getItem(
            self, "Chose country", "What's up?", ("rus", "germany", "us"), 1,  False) # ..., current, user can't edit
        if okBtnPressed:
            self.button_3.setText("Thx, bro")
            print(i)
            print(okBtnPressed)

    def run3(self):
        color = QColorDialog.getColor()
        if color.isValid():
            self.button_4.setStyleSheet(
                "background-color: {}".format(color.name()))
            print(color.name())

            
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())
```
