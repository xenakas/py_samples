```python
import pygame
import os
import numpy as np
```

```python
def load_image(name, color_key=None):
    fullname = os.path.join('data', name)
    try:
        image = pygame.image.load(fullname).convert()
    except pygame.error as message:
        print('Cannot load image:', name)
        raise SystemExit(message)
    if color_key is not None:
        if color_key == -1:
            color_key = image.get_at((0, 0))
        image.set_colorkey(color_key)
    else:
        image = image.convert_alpha()  
    return image
```

### Arrow

```python
pygame.init()

size = (500, 500)
screen = pygame.display.set_mode(size)
running = True
image = load_image("arrow.png")
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEMOTION:
            pos_move = np.array(event.pos)
    pygame.mouse.set_visible(False)
    screen.fill((0, 0, 0))
    if pygame.mouse.get_focused():
        screen.blit(image, (pos_move[0], pos_move[1]))
    pygame.display.flip()
    

pygame.quit()
```

### Creature

```python
pygame.init()

size = (300, 300)
screen = pygame.display.set_mode(size)
running = True
image = load_image("creature.png")
pos_move = np.array([0,0])

image = pygame.transform.scale(image, (110, 120))

image2 = pygame.Surface([300, 10])
image2.fill(pygame.Color("green"))


while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                pos_move[0] -= 10
            if event.key == pygame.K_RIGHT:
                pos_move[0] += 10
            if event.key == pygame.K_DOWN:
                pos_move[1] += 10
            if event.key == pygame.K_UP:
                pos_move[1] -= 10
    pygame.mouse.set_visible(False)
    screen.fill((255, 255, 255))
    
    screen.blit(image2, (0,250))
    screen.blit(image, (pos_move[0],pos_move[1]))
    
    pygame.display.flip()
    

pygame.quit()
```

```python
pygame.init()

size = (300, 300)
screen = pygame.display.set_mode(size)
running = True
pos_move = np.array([0,0])

# создадим группу, содержащую все спрайты
all_sprites = pygame.sprite.Group()
# создадим спрайт
sprite = pygame.sprite.Sprite()
# определим его вид и размеры

sprite.image = load_image("bomb.png")
sprite.rect = sprite.image.get_rect()


sprite.rect.x = 50
sprite.rect.y = 50

# добавим спрайт в группу
all_sprites.add(sprite)


while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEMOTION:
            pos_move = np.array(event.pos)
    pygame.mouse.set_visible(False)
    screen.fill((255, 255, 255))
    all_sprites.draw(screen)
    pygame.display.flip()
    
pygame.quit()
```

```python
pygame.init()

width, height = 300, 300
size = (width, height)
screen = pygame.display.set_mode(size)
running = True
pos_move = np.array([0,0])

class Bomb(pygame.sprite.Sprite):
    image = load_image("bomb.png")
    image_boom =  load_image("boom.png")

    def __init__(self, group):
        super().__init__(group)  # НЕОБХОДИМО вызвать конструктор родительского класса Sprite.
        self.image = Bomb.image
        self.rect = self.image.get_rect()
        self.rect.x = random.randrange(width-50)
        self.rect.y = random.randrange(height-50)
 
    def update(self, event):
        self.get_event(event)
 
    # Поручим бомбочке получать событие и взрываться (поменяем картинку)
    def get_event(self, event):
        for eve in event:
            if eve.type == pygame.MOUSEBUTTONDOWN:
                if self.rect.collidepoint(eve.pos):
                    self.image = Bomb.image_boom
        

# создадим группу, содержащую все спрайты
all_sprites = pygame.sprite.Group()

for _ in range(50):
    Bomb(all_sprites)

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEMOTION:
            pos_move = np.array(event.pos)
    screen.fill((0, 0, 0))
    all_sprites.draw(screen)
    all_sprites.update(pygame.event.get())
    pygame.display.flip()
    
pygame.quit()
```
